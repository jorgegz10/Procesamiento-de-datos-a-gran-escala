<h2 id="func">Funciones</h2>

Una funcin es un bloque de cdigo reutilizable que realiza las operaciones especificadas en la funcin. Le permiten dividir las tareas y reutilizar su cdigo en diferentes programas.

Hay dos tipos de funciones:

<h3 id="content">Qu es una funcin?</h3>

Puede definir funciones para proporcionar la funcionalidad requerida. Aqu hay reglas simples para definir una funcin en Python:
- Los bloques de funciones comienzan <code> def </code> seguidos por la funcin <code> name </code> y entre parntesis <code> () </code>.
- Hay parmetros de entrada o argumentos que deben colocarse entre estos parntesis.
- Tambin puede definir parmetros dentro de estos parntesis.
- Hay un cuerpo dentro de cada funcin que comienza con dos puntos (<code>: </code>) y tiene sangra.
- Tambin puede colocar la documentacin antes del cuerpo.
- La instruccin <code> return </code> sale de una funcin y, opcionalmente, devuelve un valor.

Un ejemplo de una funcin que se agrega al parmetro <code> a </code> imprime y devuelve la salida como <code> b </code>:

```python
# Primer ejemplo de funcion:  Suma 1 en a y almacena en b

def sum(a):
    b = a + 1
    print(a, "si sumas 1", b)
    return(b)
```

<h2 id="global">Variable globales</h2>

Hasta ahora, hemos estado creando variables dentro de funciones, pero no hemos discutido variables fuera de la funcin. Se denominan variables globales.
<br>
Intentemos ver qu devuelve <code> printer1 </code>:

```python
# Example of global variable

artist = "Michael Jackson"
def printer1(artist):
    internal_var = artist
    print(artist, "is an artist")

printer1(artist)
```

Si imprimimos <code> internal_var </code> obtenemos un error.

<b> Tenemos un error de nombre: <code> nombre 'internal_var' no est definido </code>. Por qu? </b>

Es porque todas las variables que creamos en la funcin es una <b> variable local </b>, lo que significa que la asignacin de la variable no persiste fuera de la funcin.

Pero hay una forma de crear <b> variables globales </b> desde dentro de una funcin de la siguiente manera:

```python
artist = "Michael Jackson"

def printer(artist):
    global internal_var
    internal_var= "Whitney Houston"
    print(artist,"is an artist")

printer(artist)
printer(internal_var)
```

<h2 id="scope">mbito de una variable</h2>

El alcance de una variable es la parte de ese programa donde esa variable es accesible. Las variables que se declaran fuera de todas las definiciones de funciones, como la variable <code> myFavouriteBand </code> en el cdigo que se muestra aqu, son accesibles desde cualquier lugar dentro del programa. Como resultado, se dice que tales variables tienen alcance global y se conocen como variables globales.
     <code> myFavouriteBand </code> es una variable global, por lo que es accesible desde la funcin <code> getBandRating </code>, y podemos usarla para determinar la clasificacin de una banda. Tambin podemos usarlo fuera de la funcin, como cuando lo pasamos a la funcin de impresin para mostrarlo:

```python
# Example of global variable

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:", getBandRating("AC/DC"))
print("Deep Purple's rating is:",getBandRating("Deep Purple"))
print("My favourite band is:", myFavouriteBand)
```

Eche un vistazo a esta versin modificada de nuestro cdigo. Ahora la variable <code> myFavouriteBand </code> est definida dentro de la funcin <code> getBandRating </code>. Una variable que se define dentro de una funcin se dice que es una variable local de esa funcin. Eso significa que solo es accesible desde dentro de la funcin en la que est definido. Nuestra funcin <code> getBandRating </code> seguir funcionando, porque <code> myFavouriteBand </code> todava est definida dentro de la funcin. Sin embargo, ya no podemos imprimir <code> myFavouriteBand </code> fuera de nuestra funcin, porque es una variable local de nuestra funcin <code> getBandRating </code>; solo se define dentro de la funcin <code> getBandRating </code>:

```python
# Example of local variable

def getBandRating(bandname):
    myFavouriteBand = "AC/DC"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is: ", getBandRating("AC/DC"))
print("Deep Purple's rating is: ", getBandRating("Deep Purple"))
print("My favourite band is", myFavouriteBand)
```

Finalmente, eche un vistazo a este ejemplo. Ahora tenemos dos definiciones de variable <code> myFavouriteBand </code>. El primero de ellos tiene un alcance global y el segundo es una variable local dentro de la funcin <code> getBandRating </code>. Dentro de la funcin <code> getBandRating </code>, la variable local tiene prioridad. ** Deep Purple ** recibir una calificacin de 10.0 cuando se pase a la funcin <code> getBandRating </code>. Sin embargo, fuera de la funcin <code> getBandRating </code>, la variable local <code> getBandRating </code> no est definida, por lo que la variable <code> myFavouriteBand </code> que imprimimos es la variable global, que tiene un valor de ** AC / DC **:

```python
# Example of global variable and local variable with the same name

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    myFavouriteBand = "Deep Purple"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:",getBandRating("AC/DC"))
print("Deep Purple's rating is: ",getBandRating("Deep Purple"))
print("My favourite band is:",myFavouriteBand)
```
