<h2 id="tuple">Tuplas</h2>

En Python existen diferentes tipos de datos: cadenas, nmeros enteros o decimales (float), etc. Todos estos tipos de datos pueden ser almacenados en una tupla como se indica a continuacin:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesType.png" width="750" align="center" />

Vamos a crear tu primera tupla compuesta de: una cadena, un entero y un nmero decimal.

```python
# Crear tu primera tupla

tuple1=(1,"2",3.0)
tuple1
```

El tipo de dato almacenado en la variable es **tuple**.

```python
# Imprime el tipo de dato de la tupla

type(tuple1)
```

<h3 id="index">Indexacin</h3>

Cada elemento dentro de la tupla puede ser accedido a travs de su nmero de ndice. La siguiente tabla representa la relacin entre el nmero de ndice y los elementos dentro de la tupla. Se puede obtener cada elemento escribiendo el nombre de la tupla seguido de corchetes conteniendo el nmero del ndice:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesIndex.gif" width="750" align="center">

Podemos imprimir cada uno de los valores de la tupla:

```python
# Imprimir la variable en cada ndice

print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
```

Podemos imprimir el **tipo** de cada uno de los valores de la tupla:

```python
# Imprimir el tipo de valor de cada uno de los ndices

print(type(tuple1[0]))
print(type(tuple1[1]))
print(type(tuple1[2]))
```

Tambin podemos usar nmeros de ndice negativos. Vamos a usar la misma tabla de arriba con su correspondiente valor negativo:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNeg.png" width="750" align="center">

Obtenemos el ultimo elemento de la siguiente forma (en esta ocasin no usaremos la sentencia print para mostrar los valores en pantalla):

```python
# Usar un valor negativo en el ndice para obtener el valor del ltimo elemento

tuple1[-1]
```

Podemos mostrar en pantalla los siguientes dos elementos de la siguiente forma:

```python
# Usar un valor negativo en el ndice para obtener el valor del penltimo elemento

tuple1[-2]
```

```python
# Usar un valor negativo en el ndice para obtener el valor del antepenltimo elemento

tuple1[-3]
```

Podemos aplicar el slice en las tuplas, obteniendo su vez nuevas tuplas con sus elementos correspondientes:

Podemos obtener los ltimos dos elementos de una tupla:

```python
# Sublista(slice) desde el ndice 3 al 4

tuple2[3:5]
```

Podemos obtener la longitud de una tupla usando el comando length:

```python
# Obtener la longitud de la tupla

len(tuple2)
```

<h3 id="nest">Tupla Anidada</h3>

Una tupla puede contener a otras tuplas, adi como otro tipo de datos complejos. Este proceso es conocido como anidar. Considere la siguiente tupla con algunos elementos:

```python
# Crear una tupla anidada

NestedT =(1, 2, ("pop", "rock") ,(3,4),("disco",(1,2)))
print(NestedT)
```

El valor de cada elemento en la tupla incluyendo a otras tuplas pueden ser obtenido a travs de un numero de ndice como se muestra en la imagen:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNestOne.png" width="750" align="center">

```python
# Imprimir el elemento de cada ndice

print("Element 0 of Tuple: ", NestedT[0])
print("Element 1 of Tuple: ", NestedT[1])
print("Element 2 of Tuple: ", NestedT[2])
print("Element 3 of Tuple: ", NestedT[3])
print("Element 4 of Tuple: ", NestedT[4])
```

Podemos usar un segundo ndice para acceder otras tuplas como se muestra en la siguiente imagen:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNestTwo.png" width="750" align="center">

Podemos acceder a las tuplas anidadas:

```python
# Imprimir el elemento de cada ndice, incluyendo los ndices anidados

print("Element 2, 0 of Tuple: ",   NestedT[2][0])
print("Element 2, 1 of Tuple: ",   NestedT[2][1])
print("Element 3, 0 of Tuple: ",   NestedT[3][0])
print("Element 3, 1 of Tuple: ",   NestedT[3][1])
print("Element 4, 0 of Tuple: ",   NestedT[4][0])
print("Element 4, 1 of Tuple: ",   NestedT[4][1])
```

Podemos acceder a las cadenas de las segundas tuplas anidadas usando un tercer nmero de ndice:

```python
# Imprimir el primer elemento en las segundas tuplas anidadas

NestedT[2][1][0]
```

```python
# Imprimir el segundo elemento en las segundas tuplas anidadas

NestedT[2][1][1]
```

Podemos usar una estructura de rbol para visualizar el proceso. Cada ndice nuevo corresponde a un nivel mas profundo del rbol:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNestThree.gif" width="750" align="center">

De igual forma, podemos acceder a elementos anidados en niveles mas bajos dentro del rbol utilizando un cuarto ndice:

```python
# Imprimir el primer elemento en las segundas tuplas anidadas

NestedT[4][1][0]
```

```python
# Imprimir el segundo elemento en las segundas tuplas anidadas

NestedT[4][1][1]
```

La siguiente figura muestra la relacin entre la estructura de rbol y el elemento <code>NestedT[4][1][1]</code>:
