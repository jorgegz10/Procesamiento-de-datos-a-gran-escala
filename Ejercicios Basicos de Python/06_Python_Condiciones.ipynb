<h3 id="comp">Operadores de Comparacin</h3>

Los operadores de comparacion relacionan algun valor u operador y, en base a una condicion, se obtiene un valor verdadero o falso. Al comparar dos valores se pueden utilizar estos operadores:

<ul>
    <li>igual: <b>==</b></li>
    <li>diferente: <b>!=</b></li>
    <li>mayor que: <b>></b></li>
    <li>menor que: <b>&lt;</b></li>
    <li>mayor o igual a: <b>>=</b></li>
    <li>menor o igual a: <b>&lt;=</b></li>
</ul>

Asignemos a la variable <code>a</code> un valor de 5. Utiliza el operador de igualdad representado por el signo <b>==</b> para determinar si dos valores son iguales. El siguiente caso compara a la variable <code>a</code> con 6.

```python
# Condicion Igual a

a = 5
a == 6
```

El resultado es <b>Falso</b> ya que 5 no es igual a 6.

Considera el siguiente operador de comparacin, <code>i > 5</code>. Si el valor del operando de la izquierda, en este caso la variable <b>i</b>, es mayor que el valor del operando de la derecha, en este caso 5, entonces la declaracin es <b>Verdadera</b>. De otra manera seria <b>Falso</b>. Si <b>i</b> es igual a 6, debido a que 6 es mayor a 5, la salida ser <b>Verdadera</b>.

```python
# Signo Mayor a

i = 6
i > 5
```

Asignar <code>i = 2</code>. La declaracin es falsa porque 2 no es mayor que 5.

```python
# Signo Mayor a

i = 2
i > 5
```

Mostraremos en la siguiente figura algunos valores para la variable <code>i</code>. Los valores mayores a 5 estn en verde y el resto en rojo. La regin verde representa a la condicin cuando esta es **Verdadera**, la roja cuando es **Falsa**. Si el valor de <code>i</code> es 2, obtenemos **Falso** ya que 2 se encuentra en la regin roja. De igual forma, si el valor de <code>i</code> es 6, tendremos **Verdadero** al estar el valor dentro del rea verde.

La comprobacin de desigualdad se representa con el signo de exclamacin precediendo el signo de igualdad, si dos operandos no son iguales entonces la condicin es **Verdadera**. Por ejemplo, la siguiente condicin producir un resultado **Verdadero** mientras el valor de <code>i</code> se distinto de 6.

```python
# Signo de Desigualdad

i = 2
i != 6
```

Cuando <code>i</code> es igual a 6 la expresin de desigualdad arroja <b>Falso</b>.

```python
# Signo de Desigualdad

i = 6
i != 6
```

Podemos aplicar el mismo mtodo a las cadenas. Por ejemplo, usando el operador de igualdad en dos de ellas. Si las cadenas no son iguales, el resultado es **Falso**.

```python
# Usar el operador de Igualdad para comparar las cadenas

"ACDC" == "Michael Jackson"
```

Si usamos el operador de desigualdad, la salida ser **Verdadera** al no ser las cadenas iguales.

```python
# Usar el operador de Desigualdad para comparar las cadenas

"ACDC" != "Michael Jackson"
```

El operador de desigualdad tambin se puede usar para comparar letras/palabras/smbolos de acuerdo con el valor ASCII de las letras. El valor decimal mostrado en la siguiente tabla representa el orden del carcter:

Por ejemplo, el cdigo ASCII de <b>!</b> es 21, mientras el cdigo para <b>+</b> es 43. Por consiguiente <b>+</b> es mayor que <b>!</b> ya que 43 es mayor a 21.

De igual forma, el valor ASCII de <b>A</b> es 101, y el valor para <b>B</b> es 102, entonces:

<h3 id="branch">Sentencias Condicionales (Branching)</h3>

Las sentencias condicionales permiten ejecutar diferentes declaraciones para distintas entradas. Es til pensar en una **condicin if** como una habitacin cerrada, si la condicin es **Verdadera** podremos entrar y la aplicacin podr ejecutar algunas tareas predefinidas, pero si la condicin es **Falsa** el programa ignorara dichas tareas.

Por ejemplo, consideremos que el rectngulo azul representa un concierto de ACDC. Si la persona es mayor de 18, puede entrar al concierto. Si tiene 18 o menos, no puede entrar.

Utiliza la sentencia condicional vista anteriormente como aquella necesaria para usarse en una **condicin if**. La sintaxis es tan sencilla como <code> if <i>sentencia condicional</i> :</code>, la cual contiene la palabra <code>if</code>, la condicin y dos puntos al final. Empieza la tarea que tienes que ejecutar con esta condicin en una nueva lnea y sangra. Las lneas de cdigo despus de los dos puntos y con sangra solo sern ejecutadas cuando la **condicin if** sea **Verdadera**. Las tareas finalizaran cuando el cdigo no contenga sangra.

En el siguiente caso, las tareas sole ejecutaran <code>print(you can enter)</code> si la variable <code>age</code> es mayor a 18 es Verdadera ya que esta parte del cdigo contiene sangra. Sin embargo, la ejecucin de <code>print(move on)</code> no ser afectada por la condicin if.

```python
# Ejemplo de condicion if

age = 19
#age = 18

#La expresin que puede ser verdadera o falsa
if age > 18:

    #parte con sangra, esta es la expresin que ser ejecutada si la condicin es verdadera
    print("you can enter" )

#Las declaraciones despus de la condicin if sern ejecutadas independientemente de si la condicin
#es verdadera o falsa
print("move on")
```

La condicin <code>else</code> ejecuta un bloque de cdigo si ninguna de las condiciones antes de ella es **Verdadera**.  Usemos de nuevo la analoga del concierto de ACDC. Si la persona tiene 17 aos no puede entrar al concierto de ACDC, pero si puede entrar al de MeatLoaf. La sintaxis de la condicin <code>else</code> es similar a la de la condicin <code>if</code>, de esta forma <code>else :</code>. Observa que no hay una declaracin de condicin para <code>else :</code>.
```python
# Ejemplo de condicion Else

age = 18
# age = 19

if age > 18:
    print("you can enter" )
else:
    print("go see Meat Loaf" )

print("move on")
```

La declaracin <code>elif</code>, una abreviacin de else if, nos permite comprobar condiciones adicionales si otras antes de ella son <b>Falsas</b>. Si la condicin de la declaracin <code>elif</code> es <b>Verdadera</b>, la expresin alternativa ser ejecutada. Considera el ejemplo del concierto, donde si la persona tiene 18 ira al concierto de Pink Floyd en lugar de ACDC o MeatLoaf. La persona de 18 entra, debido a que no es mayor a 18 no puede ver a ACDC, pero puede ver a Pink Floyd. Despus de ver el concierto, prosigue. La sintaxis de la declaracin <code>elif</code> es similar a la de <code>if</code> ya que apenas cambiamos la declaracin <code>if</code> por <code>elif</code>.

```python
# Ejemplo de declaracin elif

age = 18

if age > 18:
    print("you can enter" )
elif age == 18:
    print("go see Pink Floyd")
else:
    print("go see Meat Loaf" )

print("move on")
```

<h3 id="logic">Operadores Lgicos</h3>

A veces es necesario revisar mas de una condicin al mismo tiempo. Por ejemplo, pudieras querer ver si una condicin y otra son **Verdaderas**. Los operadores lgicos permiten combinar o modificar condiciones.
<ul>
    <li><code>and</code></li>
    <li><code>or</code></li>
    <li><code>not</code></li>
</ul>

Estos el funcionamiento de estos operadores se muestra en la siguiente tabla de verdad:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/CondsTable.png" width="650" />

La declaracin <code>and</code> solo es **Verdadera** cuando ambas condiciones lo son. En el caso de <code>or</code> ser verdadero si alguna de las condiciones es **Verdadera**. La declaracin <code>not</code> tendr como resultado el valor de verdad opuesto a la salida.

Veamos ahora como determinar si un disco musical fue lanzado despus de 1979 (sin incluir ese ao) y antes de 1990 (sin incluir ese ao). El periodo 1980 y 1989 satisface esta condicin. Se demostrar esto en la siguiente figura. La parte verde de las lneas <strong>a</strong> y <strong>b</strong> representa periodos donde la declaracin es **Verdadera**. La parte verde en la lnea <strong>c</strong> representa el periodo donde la lnea a y b son **Verdaderas** al mismo tiempo.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/CondsEgOne.png" width="650" />

El bloque de cdigo que implementa esto es:

```python
# Condicin de ejemplo

album_year = 1980

if(album_year > 1979) and (album_year < 1990):
    print ("Album year was in between 1980 and 1989")

print("")
print("Do Stuff..")
```

Para determinar si un disco fue lanzado antes de 1980 (.. - 1979) o despus de 1989 (1990 - .. ) se puede utilizar una declaracin **or**. Los periodos anteriores a 1980 (.. - 1979) o posterior a 1989 (1990 - ..) satisfacen esta condicin. Esto se muestra en la figura siguiente, el color verde en la lnea <strong>a</strong> y <strong>b</strong> representa periodos donde la declaracin tiene un valor verdadero. El color verde en **c** indica que la condicin es verdadera en al menos una de las condiciones.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/CondsEgTwo.png" width="650" />

El bloque de cdigo que implementa esto es:

```python
# Condicin de ejemplo

album_year = 1990

if(album_year < 1980) or (album_year > 1989):
    print ("Album was not made in the 1980's")
else:
    print("The Album was made in the 1980's ")
```

La declaracin <code>not</code> revisa si la condicin es falsa:

```python
# Condicin de ejemplo

album_year = 1983

if not (album_year == '1984'):
    print ("Album year is not 1984")
```
