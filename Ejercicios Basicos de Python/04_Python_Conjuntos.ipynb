<h3 id="content">Contenido de un Conjunto</h3>

Un conjunto es una coleccin de objetos nicos en Python. Un conjunto se define usando llaves de apertura y cierre <b>{}</b>. Python remover automticamente elementos duplicados.

```python
# Crear un conjunto

set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
set1
```

Tambin se puede crear un conjunto a partir de una lista:

```python
# Convertir una lista en un conjunto

album_list = [ "Michael Jackson", "Thriller", 1982, "00:42:19", \
              "Pop, Rock, R&B", 46.0, 65, "30-Nov-82", None, 10.0]
album_set = set(album_list)
album_set
```

<h3 id="op">Operaciones con Conjuntos</h3>

Vamos a revisar las operaciones con conjuntos que nos ayudaran a modificarlo. Considera el conjunto <b>A</b>:

```python
# Conjunto de ejemplo

A = set(["Thriller", "Back in Black", "AC/DC"])
A
```

Podemos agregar un elemento al conjunto mediante el mtodo <code>add()</code>:

```python
# Agregar un elemento al conjunto

A.add("NSYNC")
A
```

Si agregamos el mismo elemento dos veces nada cambiara debido a que no puede haber elemento duplicados en un conjunto:

```python
# Trata de agregar un elemento duplicado al conjunto

A.add("NSYNC")
A
```

Podemos borrar un elemento del conjunto mediante el mtodo <code>remove</code>:

```python
# Remover un elemento del conjunto

A.remove("NSYNC")
A
```

Podemos verificar si existe un elemento especifico dentro del conjunto con el comando <code>in</code>:

```python
# Verificar si existe un elemento en el conjunto

"AC/DC" in A
```

Recuerda que puedes revisar la diferencia entre dos conjuntos, as como la diferencia simtrica, interseccin y unin:

Considera estos dos conjuntos:

```python
# Conjuntos de ejemplo

album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
```

Como los dos conjuntos contienen <b>AC/DC</b> y <b>Back in Black</b> podemos representar estos elementos en comn en la interseccin de dos crculos.

Para encontrar la interseccin de dos conjuntos usaremos <code>&</code>:

```python
# Encontrar las intersecciones

intersection = album_set1 & album_set2
intersection
```

Puedes encontrar los elementos que nicamente estn dentro de <code>album_set1</code> usando el mtodo <code>difference</code>:

```python
# Encuentra la diferencia en set1 pero no en set2

album_set1.difference(album_set2)
```

Solo tienes que considerar los elementos en <code>album_set1</code>; todos los de <code>album_set2</code>, incluyendo los de la interseccin, no estn incluidos.

```python
# Usar el mtodo intersection para encontrar la interseccin entre of album_list1 y album_list2

album_set1.intersection(album_set2)
```

La unin es la suma de todos los elementos de ambos conjuntos, se representa pintando los dos crculos:

La unin esta dada por:

```python
# Encontrar la unin de los dos conjuntos

album_set1.union(album_set2)
```

Se puede averiguar si un conjunto es un sper conjunto o un sub conjunto de otro, respectivamente, de la siguiente forma:

```python
# Revisar si es un super conjunto

set(album_set1).issuperset(album_set2)
```

```python
# Revisar si es un sub conjunto

set(album_set2).issubset(album_set1)
```

Este es un ejemplo donde <code>issubset()</code> y <code>issuperset()</code> regresan un valor verdadero:

```python
# Revisar si es un sub conjunto

set({"Back in Black", "AC/DC"}).issubset(album_set1)
```

```python
# Revisar si es un super conjunto

album_set1.issuperset({"Back in Black", "AC/DC"})
```
